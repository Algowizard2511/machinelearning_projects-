```python
import pickle
import numpy as np
```

```python
enc=pickle.load(open('/content/enc.pkl','rb'))
ohe=pickle.load(open('/content/ohe.pkl','rb'))
scaler=pickle.load(open('/content/scaler.pkl','rb'))
clf=pickle.load(open('/content/clf.pkl','rb'))
```

```python
#pclass	sex	age	sibsp	parch	fare	embarked	deck
user_input=np.array([2,'male',28,1,2,200,'S','C'],dtype='object').reshape(1,8)
```

```python
user_input
```

```python
#ordinal encoding
#enc.transform(user_input[:,0].reshape(1,1))
```

```python
#one hot encoding
hot_input=ohe.transform(user_input[:,[1,6,7]].reshape(1,-1))
hot_input.shape
```

```python
#standardization
std_col=scaler.transform(user_input[:,[2,5]])
std_col.shape
```

```python
new_input=np.concatenate([user_input[:,[0]],hot_input,std_col],axis=1)
```

```python
predicted=clf.predict(new_input)
```

```python
if predicted==1:
  print('Passenger Survived')
else:
  print('Not Survived')
```
